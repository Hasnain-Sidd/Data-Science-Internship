```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading Iris Data Set
df=pd.read_csv("C:\\Users\\PMLS\\Downloads\\archive (10)\\Iris.csv")
df
```

```python
# Checking If Data Set Contain Missing Values 
df.isnull().sum()
```

```python
# I have dropped The ID Colum because It will have no affect on the visualization
df.drop(columns={'Id'},inplace=True)
```

```python
# Shape of the data frame
df.shape
```

```python
# Columns in the data set
df.columns
```

```python
# Head Of the Data frame
df.head()
```

```python
# Scatter Plot for checking relation between Sepal Lenght and Width
plt.figure(figsize=(8, 6))
plt.scatter(df.loc[0:49, 'SepalLengthCm'], df.loc[0:49, 'SepalWidthCm'], 
            label='Setosa', color='#1f77b4', edgecolor='black', s=60, alpha=0.8)
plt.scatter(df.loc[50:99, 'SepalLengthCm'], df.loc[50:99, 'SepalWidthCm'], 
            label='Versicolor', color='#ff7f0e', edgecolor='black', s=60, alpha=0.8)
plt.scatter(df.loc[100:149, 'SepalLengthCm'], df.loc[100:149, 'SepalWidthCm'], 
            label='Virginica', color='#2ca02c', edgecolor='black', s=60, alpha=0.8)
plt.xlabel('Sepal Length (cm)', fontsize=12, fontweight='bold')
plt.ylabel('Sepal Width (cm)', fontsize=12, fontweight='bold')
plt.title('Sepal Length vs Sepal Width by Species', fontsize=15, fontweight='bold')
plt.legend(title='Species', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot for checking Relationsip between Petal Length and Width
plt.figure(figsize=(8, 6))
plt.scatter(df.loc[0:49, 'PetalLengthCm'], df.loc[0:49, 'PetalWidthCm'],
            label='Setosa', color='#1f77b4', edgecolor='black', s=60, alpha=0.8)
plt.scatter(df.loc[50:99, 'PetalLengthCm'], df.loc[50:99, 'PetalWidthCm'],
            label='Versicolor', color='#ff7f0e', edgecolor='black', s=60, alpha=0.8)
plt.scatter(df.loc[100:149, 'PetalLengthCm'], df.loc[100:149, 'PetalWidthCm'],
            label='Virginica', color='#2ca02c', edgecolor='black', s=60, alpha=0.8)
plt.xlabel('Petal Length (cm)', fontsize=12, fontweight='bold')
plt.ylabel('Petal Width (cm)', fontsize=12, fontweight='bold')
plt.title('Petal Length vs Petal Width by Species', fontsize=15, fontweight='bold')
plt.legend(title='Species', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Distribution Of Iris Data Set Features
sns.set()
ax = df.hist(
    figsize=(10, 8),
    color='#4C72B0',
    edgecolor='black',
    grid=False
)
plt.suptitle('Distribution of Iris Dataset Features', fontsize=18, fontweight='bold', color='#333333')
plt.tight_layout(rect=[0, 0, 1, 0.96])
for axs in ax.flatten():
    axs.set_facecolor('lightblue')
    axs.title.set_fontsize(14)
    axs.title.set_color('#4C72B0')
    axs.set_xlabel(axs.get_xlabel(), fontsize=12)
    axs.set_ylabel(axs.get_ylabel(), fontsize=12)
plt.show()

```

```python
# Total Count of each specie in Our Data Set
ax=df['Species'].value_counts().plot(kind='bar',edgecolor='black',width=.5)
sns.set()
for bars in ax.containers:
    ax.bar_label(bars,fontsize=8,fontweight='bold')
ax.set_facecolor('lightblue')
plt.xlabel("Species",fontsize=12,fontweight='bold')
plt.ylabel("Count",fontsize=12,fontweight='bold')
plt.xticks(rotation=0,fontweight='bold')
plt.title("Bar Chart Of Species Distribution",fontweight='bold',fontsize=15)
```

```python
# Boxplot for all numerical features in the Iris dataset with enhanced style
sns.set()
ax=sns.boxplot(data=df,palette='dark')
plt.title('Boxplot of Iris Dataset Features', fontsize=15, fontweight='bold', color='#4C72B0')
plt.ylabel('cm', fontsize=12, color='#333333')
plt.yticks(fontsize=11, color='#333333')
plt.gca().set_facecolor('whitesmoke')
plt.xticks(color='#4C72B0',fontsize=12,fontweight='bold')
plt.tight_layout()
plt.show()
```
